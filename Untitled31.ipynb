```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
categories = ['Net Patient Revenue', 'Other Revenue', 'Total Operating Revenue',
              'Physician Salaries', 'Physician Incentive Compensation', 'Nursing Salaries',
              'Administrative Salaries', 'Benefit Expenses', 'Total Compensation Expenses',
              'Medical Supply Expense', 'Pharmacy Expense', 'Clerical Supply Expense',
              'Depreciation Expense', 'Rent Expense', 'Other Expense',
              'Total General & Administrative Expenses', 'Total Operating Expenses',
              'Excess (Loss) of Revenue over Expenses']

actual_values = [7588476, 78884, 7667360,
                 2833027, 1448510, 1911473,
                 210000, 704331, 7107341,
                 163200, 200462, 12407,
                 22000, 90000, 40220,
                 528289, 7635630, 31730]

budget_values = [7500000, 80000, 7580000,
                 3000000, 1250000, 1750000,
                 225000, 684750, 6909750,
                 150000, 200000, 15000,
                 22000, 90000, 50000,
                 527000, 7436750, 143250]

# Create DataFrame
df = pd.DataFrame({'Category': categories, 'Actual': actual_values, 'Budget': budget_values})

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df['Category'], df['Actual'], marker='o', label='Actual')
plt.plot(df['Category'], df['Budget'], marker='o', linestyle='--', color='red', label='Budget')
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Amount ($)')
plt.title('Actual vs. Budgeted Values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()



```

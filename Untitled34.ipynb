```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

#  Loading the MNIST dataset
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

#  Preprocessing the data
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train = x_train[..., tf.newaxis]
x_test = x_test[..., tf.newaxis]

#  Defining different CNN architectures
def create_cnn_model(num_layers, num_filters):
    model = models.Sequential()
    model.add(layers.Conv2D(num_filters, (3, 3), activation='relu', input_shape=(28, 28, 1)))
    for _ in range(num_layers - 1):
        model.add(layers.Conv2D(num_filters, (3, 3), activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(10, activation='softmax'))
    return model

# Step 4: Train and evaluate each CNN model
def train_and_evaluate_model(model, epochs=5):
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs=epochs, validation_data=(x_test, y_test))
    return history

#  Comparing the results
def plot_history(history, label):
    plt.plot(history.history['accuracy'], label='accuracy')
    plt.plot(history.history['val_accuracy'], label='val_accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Model Accuracy - ' + label)
    plt.legend()
    plt.show()

#  CNN with 2 layers and 32 filters
model_1 = create_cnn_model(num_layers=2, num_filters=32)
history_1 = train_and_evaluate_model(model_1)
plot_history(history_1, label='CNN 2 layers, 32 filters')

# CNN with 4 layers and 64 filters
model_2 = create_cnn_model(num_layers=4, num_filters=64)
history_2 = train_and_evaluate_model(model_2)
plot_history(history_2, label='CNN 4 layers, 64 filters')

# CNN with 6 layers and 128 filters
model_3 = create_cnn_model(num_layers=6, num_filters=128)
history_3 = train_and_evaluate_model(model_3)
plot_history(history_3, label='CNN 6 layers, 128 filters')




```

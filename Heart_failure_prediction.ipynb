```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load the dataset
file_path = '/content/Preprocesed dataset FINAL PROJECT.xlsx'
df = pd.read_excel(file_path)

# Display descriptive statistics
descriptive_stats = df.describe()
print("Descriptive Statistics:")
print(descriptive_stats)

# Reliability Analysis (if applicable - you need to specify the relevant variables)
# Example: reliability_analysis = df[['variable1', 'variable2']].corr()

# PCA Analysis
X = df.drop('HeartDisease', axis=1)
y = df['HeartDisease']

# Standardize the features
X_standardized = StandardScaler().fit_transform(X)

# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_standardized)

# Display explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("\nExplained Variance Ratio:")
print(explained_variance_ratio)

# Data Normalization
# (If you have variables that need normalization, you can use Min-Max scaling or other methods)

# Machine Learning Models
X_train, X_test, y_train, y_test = train_test_split(X_standardized, y, test_size=0.2, random_state=42)

# Support Vector Machine (SVM)
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)
svm_report = classification_report(y_test, svm_predictions)

# Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_report = classification_report(y_test, rf_predictions)

# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_predictions)
lr_report = classification_report(y_test, lr_predictions)

# Gradient Boosting
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)
gb_predictions = gb_model.predict(X_test)
gb_accuracy = accuracy_score(y_test, gb_predictions)
gb_report = classification_report(y_test, gb_predictions)

# Neural Network
nn_model = MLPClassifier()
nn_model.fit(X_train, y_train)
nn_predictions = nn_model.predict(X_test)
nn_accuracy = accuracy_score(y_test, nn_predictions)
nn_report = classification_report(y_test, nn_predictions)

# Display evaluation metrics
print("\nEvaluation Metrics:")
print("SVM Accuracy:", svm_accuracy)
print("Random Forest Accuracy:", rf_accuracy)
print("Logistic Regression Accuracy:", lr_accuracy)
print("Gradient Boosting Accuracy:", gb_accuracy)
print("Neural Network Accuracy:", nn_accuracy)

print("\nSVM Report:")
print(svm_report)

print("\nRandom Forest Report:")
print(rf_report)

print("\nLogistic Regression Report:")
print(lr_report)

print("\nGradient Boosting Report:")
print(gb_report)

print("\nNeural Network Report:")
print(nn_report)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
file_path = 'Preprocesed dataset FINAL PROJECT.xlsx'  # Adjust the path accordingly
df = pd.read_excel(file_path)

# Separate numerical features (assumes all features except 'HeartDisease' are numerical)
numerical_features = df.drop('HeartDisease', axis=1)

# Standardize numerical features for PCA
numerical_features_standardized = StandardScaler().fit_transform(numerical_features)

# Perform PCA
pca = PCA()
principal_components = pca.fit_transform(numerical_features_standardized)

# Explained Variance Ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:")
print(explained_variance_ratio)

# Scree Plot
singular_values = pca.singular_values_
scree_plot_data = singular_values**2 / np.sum(singular_values**2)
plt.plot(np.arange(1, len(scree_plot_data) + 1), scree_plot_data, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.title('Scree Plot')
plt.show()
```

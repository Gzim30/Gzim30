```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define nodes (libraries)
libraries = [
    'NumPy', 'Pandas', 'Matplotlib', 'Seaborn', 'Scikit-learn', 'TensorFlow', 'PyTorch', 'NLTK', 'Spacy'
]

# Add nodes to the graph
G.add_nodes_from(libraries)

# Define edges (relationships between libraries)
edges = [
    ('NumPy', 'Pandas'), ('NumPy', 'Matplotlib'), ('NumPy', 'Scikit-learn'),
    ('Pandas', 'Matplotlib'), ('Pandas', 'Seaborn'),
    ('Scikit-learn', 'TensorFlow'), ('Scikit-learn', 'PyTorch'),
    ('NLTK', 'Spacy')
]

# Add edges to the graph
G.add_edges_from(edges)

# Set node colors
node_colors = ['skyblue' if node not in ['NLTK', 'Spacy'] else 'lightcoral' for node in G.nodes]

# Set node positions
pos = nx.spring_layout(G, seed=42)
pos['Matplotlib'] = (pos['Matplotlib'][0], pos['Matplotlib'][1] + 0.1)  # Move 'Matplotlib' higher

# Draw the graph
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color=node_colors, font_size=10, font_color="black", font_weight="bold", arrowsize=20)
plt.title('Relationships Between Python Libraries', fontsize=16)
plt.show()











```
